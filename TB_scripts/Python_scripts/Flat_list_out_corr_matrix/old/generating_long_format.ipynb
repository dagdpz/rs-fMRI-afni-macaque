```python
import csv
import pandas as pd

# Define paths to input and output files
input_file = "correlation_files/sub-Linus_predrug_no_smoothing.txt"
output_file = "correlation_files/csv_tables/correlation_matrix_long_format_predrug.csv"

# Define line numbers
header_line = 4  # Line where axis labels are located
start_line = 7   # First line of the matrix (after the header line)
end_line = 14    # Last line of the matrix

# Read text file and extract relevant lines
with open(input_file, 'r') as file:
    lines = file.readlines()
    
    # Extract axis labels (header)
    header = lines[header_line - 1].strip().split()
    
    # Extract matrix lines
    matrix_lines = lines[start_line - 1:end_line]
    matrix_data = [line.strip().split() for line in matrix_lines]

# Convert the matrix to long format
long_format_data = []
for i, row in enumerate(matrix_data):
    for j, value in enumerate(row):
        long_format_data.append({
            'Variable1': header[i],
            'Variable2': header[j],
            'Correlation': float(value)  # Save correlation as a float
        })

# Convert the data into a Pandas DataFrame
long_format_df = pd.DataFrame(long_format_data)

# Save the long format as a CSV file
long_format_df.to_csv(output_file, index=False)

print(f"The correlation matrix in long format has been saved to {output_file}.")

# Postdrug
import csv
import pandas as pd

# Define paths to input and output files
input_file = "correlation_files/sub-Linus_postdrug_no_smoothing.txt"
output_file = "correlation_files/csv_tables/correlation_matrix_long_format_postdrug.csv"

# Define line numbers
header_line = 4  # Line where axis labels are located
start_line = 7   # First line of the matrix (after the header line)
end_line = 14    # Last line of the matrix

# Read text file and extract relevant lines
with open(input_file, 'r') as file:
    lines = file.readlines()
    
    # Extract axis labels (header)
    header = lines[header_line - 1].strip().split()
    
    # Extract matrix lines
    matrix_lines = lines[start_line - 1:end_line]
    matrix_data = [line.strip().split() for line in matrix_lines]

# Convert the matrix to long format
long_format_data = []
for i, row in enumerate(matrix_data):
    for j, value in enumerate(row):
        long_format_data.append({
            'Variable1': header[i],
            'Variable2': header[j],
            'Correlation': float(value)  # Save correlation as a float
        })

# Convert the data into a Pandas DataFrame
long_format_df = pd.DataFrame(long_format_data)

# Save the long format as a CSV file
long_format_df.to_csv(output_file, index=False)

print(f"The correlation matrix in long format has been saved to {output_file}.")
```
